```python
#https://platform.openai.com/docs/guides/function-calling
functions = [
        {
            "name": "get_current_weather",
            "description": "Get the current weather in a given location",
            "parameters": {
                "type": "object",
                "properties": {
                    "location": {
                        "type": "string",
                        "description": "The city and state, e.g. San Francisco, CA",
                    },
                    
                },
                "required": ["location"],
            },
        }
    ]
```

```python
user_message="What is the temperature of Bangalore"

messages.append({"role": "user", "content": user_message})
completion=client.chat.completions.create(
    model="gpt-3.5-turbo-0613",
    messages=messages,
    functions=functions
    
)
```

```python
messages
```

```python
completion
```

```python
print(completion.choices[0].message)
```

```python
response=completion.choices[0].message
```

```python
response
```

```python
function_name = response.function_call.name
print(function_name)
```

```python
import json
#location=eval(response['function_call']['arguments'])['location']
# Assuming response is an object of type ChatCompletionMessage
arguments_str = response.function_call.arguments
arguments_dict = json.loads(arguments_str)
location = arguments_dict['location']
print(location)
```

```python
# Step 4: send the info on the function call and function response to GPT
messages.append(response)  # extend conversation with assistant's reply
messages.append(
    {
        "role": "function",
        "name": function_name,
        "content": location,
    }
)
```

```python
messages
```

```python
# extend conversation with function response
second_response = client.chat.completions.create(
    model="gpt-3.5-turbo-0613",
    messages=messages,
    functions=functions
)  # get a new response from GPT where it can see the function response

```

```python
print(second_response.choices[0].message)
```
